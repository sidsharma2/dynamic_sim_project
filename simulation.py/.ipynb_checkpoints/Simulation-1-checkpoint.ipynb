#Mass-Spring-Damper System

The equation of motion for a mass-spring-damper system is:

$$
m\ddot{x} + c\dot{x} + kx = 0
$$

Where:
- \( m \) is the mass (kg)  
- \( c \) is the damping coefficient (NÂ·s/m)  
- \( k \) is the spring constant (N/m)  
- \( x \) is the displacement (m)

We will solve this using numerical integration.

```python
import numpy as np
import matplotlib.pyplot as plt

# System parameters
m = 1.0   # Mass (kg)
k = 10.0  # Spring constant (N/m)
c = 0.5   # Damping coefficient (NÂ·s/m)

# Time setup
t_max = 10  # seconds
dt = 0.01   # time step
t = np.arange(0, t_max, dt)

# Initial conditions
x = np.zeros_like(t)  # Displacement
v = np.zeros_like(t)  # Velocity
x[0] = 1.0  # Initial displacement (m)

# Euler Method for numerical integration
for i in range(1, len(t)):
    a = (-c * v[i-1] - k * x[i-1]) / m  # Acceleration
    v[i] = v[i-1] + a * dt              # Velocity update
    x[i] = x[i-1] + v[i-1] * dt        # Displacement update

# Plotting the results
plt.figure(figsize=(10, 6))
plt.plot(t, x, label='Displacement (x)')
plt.xlabel('Time (s)')
plt.ylabel('Displacement (m)')
plt.title('Mass-Spring-Damper System Simulation')
plt.legend()
plt.grid(True)
plt.show()
```

# Mass-Spring-Damper System Simulation

The dynamics of a **mass-spring-damper** system are governed by the second-order differential equation:


$$
m\ddot{x}(t) + c\dot{x}(t) + kx(t) = 0
$$


Where:  
- \( m \) â **Mass** (kg)  
- \( c \) â **Damping coefficient** (NÂ·s/m)  
- \( k \) â **Spring constant** (N/m)  
- \( x(t) \) â **Displacement** (m)  
- \( \dot{x}(t) \) â **Velocity** (m/s)  
- \( \ddot{x}(t) \) â **Acceleration** (m/sÂ²)

---

## âï¸ **Numerical Integration: Euler Method**

The system is solved numerically using the **Euler Method**:

**Velocity Update:**  
$$
v_{i} = v_{i-1} + a_{i-1} \Delta t
$$

$$
x_{i} = x_{i-1} + v_{i-1} \Delta t
$$

Where:  
- \( a_{i} = \frac{-c v_{i} - k x_{i}}{m} \) â **Acceleration**  
- \( \Delta t \) â **Time step size**

---

### ð¯ **Objective**

This simulation visualizes the **damped oscillation** of a mass-spring-damper system over time.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from ipywidgets import interact, FloatSlider
from scipy.integrate import solve_ivp

# ð Define the system dynamics
def mass_spring_damper(t, y, m, c, k, F0, omega):
    x, v = y
    dxdt = v
    dvdt = (-c * v - k * x + F0 * np.cos(omega * t)) / m
    return [dxdt, dvdt]

# ð Function to simulate and animate
def simulate_system(m=1.0, c=0.5, k=10.0, F0=0.0, omega=1.0):
    # Simulation settings
    t_span = (0, 20)
    t_eval = np.linspace(*t_span, 1000)
    
    # Solve ODE
    sol = solve_ivp(mass_spring_damper, t_span, [1.0, 0.0], t_eval=t_eval, args=(m, c, k, F0, omega))

    # Animation setup
    fig, ax = plt.subplots(figsize=(8, 4))
    ax.set_xlim(-2, 2)
    ax.set_ylim(-1.5, 1.5)
    ax.set_xlabel('Displacement (m)')
    ax.set_title('Mass-Spring-Damper System')
    mass, = ax.plot([], [], 'o-', markersize=20, label='Mass')
    spring, = ax.plot([], [], 'k-', lw=2, label='Spring')

    def init():
        mass.set_data([], [])
        spring.set_data([], [])
        return mass, spring

    def update(frame):
        x = sol.y[0, frame]
        mass.set_data(x, 0)
        spring.set_data([0, x], [0, 0])
        return mass, spring

    ani = FuncAnimation(fig, update, frames=len(t_eval), init_func=init, interval=20, blit=True)
    plt.legend()
    plt.grid(True)
    plt.show()

# ðï¸ Interactive sliders for parameter adjustment
interact(simulate_system,
         m=FloatSlider(value=1.0, min=0.1, max=5.0, step=0.1, description='Mass (kg)'),
         c=FloatSlider(value=0.5, min=0.0, max=5.0, step=0.1, description='Damping (Ns/m)'),
         k=FloatSlider(value=10.0, min=1.0, max=20.0, step=0.5, description='Spring (N/m)'),
         F0=FloatSlider(value=0.0, min=0.0, max=10.0, step=0.5, description='Force Amplitude'),
         omega=FloatSlider(value=1.0, min=0.1, max=10.0, step=0.1, description='Force Frequency'))
```
